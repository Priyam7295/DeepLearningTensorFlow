```python
import tensorflow as tf
print(tf.__version__)
```

### Creating tensors with tf.constant

```python
scalar = tf.constant(7)
print(scalar)
print(scalar.ndim)
```

```python
vector = tf.constant([10,10])
print(vector)
print(vector.ndim)
```

```python
matrix = tf.constant( [[1,2,3],[4,5,3],[2,32,1]] )
matrix2 = tf.constant( [[1,21,3],[41,5,3],[22,32,11]] )

matrix.shape ,matrix2.shape
```

```python
print(matrix@matrix2)

print()
print(tf.matmul(matrix ,matrix2))
```

```python
vector=tf.constant([-1.0,2.0])

print(vector.dtype,"\n")

b= tf.cast(vector ,tf.float16)
print(b.dtype)
```

```python
print(tf.abs(vector),"\n")
# print(vector)
```

```python
arr= np.random.randint(0,100,23)
print(arr)

print("\n")
ten = tf.constant(arr)
print(ten)
print("\n")

```

```python
# For std we need to convert our values from int to float
print(tf.math.reduce_std( tf.cast(ten ,dtype=tf.float32) ))
```

```python
#squeezing a tensor(removing all single dimensions)

# Create a tensor to get started
import numpy as np
np.random.seed(42)
G = np.random.uniform(size=50)
ten = tf.constant(G)
ten2=tf.reshape(ten,shape=(1,1,1,50) )
ten2
```

```python
G_squeezed= tf.squeeze(G)
G_squeezed
```
